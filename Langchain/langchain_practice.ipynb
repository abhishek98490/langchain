### Text Splitter

```python
# pdf loader

from langchain_community.document_loaders import PyMuPDFLoader
loader = PyMuPDFLoader("D:\\MLOps\\langchain\\langchain\\data\\Capstone Project Report Format.pdf")
doc = loader.load()
```

```python
# Recursively split the text
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100,chunk_overlap=50)
finall_text = text_splitter.split_documents(doc)
```

### Embedding Technique Using HuggingFace

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HF_TOKEN'] = os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "test document"

result = embeddings.embed_query(text) # also can give list of text for embeddings
```
